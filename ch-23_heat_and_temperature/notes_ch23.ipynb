# __CH-23 HEAT AND TEMPERATURE__

## 23.1 HOT AND COLD BODIES

__Heat__

→ When a hot body is kept in contact with a cold body, the cold body warms and the hot body cools down

→ The internal energy of the hot body decreases and the internal energy of the cold body increases

→ Thus, energy is transferred from the hot body to the cold body when they are placed in contact

→ The transfer of energy from a hot body to a cold body is a non-mechanical process

→ _The energy that is transferred from one body to the other, without any mechanical work involved is called heat_

→ as hot body cools down its internal energy decreases 

→ and when cold body warms up its internal energy increases
